```python
i=1
while i<11:
    square=i*i
    print("square is ",square)
    if square==25:
        break
    i=i+1
print("last line")
```

```python
i=1
while i<11:
    square=i*i
    if square==9:
        print("skip square 9")
        i=i+1
        continue
    print("the value of square is",square)
    i=i+1
print("last line")
```

# filter function

```python
# WAP to filter even numbers
num_list=[1,52,8,12,24,33,42,3,6,7,0]
even_function=lambda x:x%2==0
result=filter(even_function,num_list)
list(result)
```

```python
num=[4,5,8,9,0,47,87,90,23]
even=lambda y:y%2==0
result=filter(even,num)
list(result)
```

```python
# WAP to filter number greater than 100
num=[100,200,50,300,366,80,60,302,101]
greater_than_100=lambda z:z>100
result=filter(greater_than_100,num)
list(result)
```

```python
# *args
def add(*args):
    total=0
    for num in args:
        total +=num
    return total

print(add(1,3,5)) # can change any number ,include any number for addn
```

```python
#  **kwargs understands by default as key values 
def create_profile(**kwrags):
    print("User Profile")
    for key,value in kwrags.items():
        print(f"{key}:{value}")

create_profile(name='subi',age=21,job='Data Analyst')
```
