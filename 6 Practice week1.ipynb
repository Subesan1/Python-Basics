```python
# BREAK
i=1
while i<12:
    square = i ** 2
    print ('sq value is ',square)
    if square ==36:
        break 
    i=i+1
```

```python
# CONTINUE 
n=1
while n<7:
    cube = n*n*n
    if cube == 125:
        print("skip cube 125")
        n=n+1
        continue
    print("cube val is",cube)
    n=n+1
print("the end")
```
