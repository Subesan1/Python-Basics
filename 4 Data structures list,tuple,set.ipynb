## set operations

```python
A={1,2,3,4,5}
B={4,5,6,7,8}
```

```python
A.union(B)
```

```python
A.intersection(B)
```

```python
A.difference(B) # returns all the elements only present in A
```

```python
B.difference(A) # returns all the elements only present in B
```
