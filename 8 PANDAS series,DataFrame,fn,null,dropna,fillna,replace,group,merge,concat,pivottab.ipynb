# merge

```python
df1=pd.DataFrame({
    'roll_no':[1,2,3,4,5],'arts':[70,80,76,53,73]
 })

df2=pd.DataFrame({
    'roll_no':[1,2,3,4,5],'science':[45,32,65,32,21]
})
# to show it is in this order 
```

```python
merged_df=pd.merge(df1,df2, on='roll_no') # if not mentioned from on it automatically checks common values and assigns
print(merged_df)
```

```python
merged_df=pd.merge(df1,df2)
print(merged_df)
```

```python
merged_df=pd.merge(df2,df1) # swaps
print(merged_df)
```

```python
df3=pd.DataFrame({
    'row_no':[1,2,3,6,7],'arts_stu':[70,80,76,53,73]
 })

df4=pd.DataFrame({
    'row_no':[1,2,3,8,5],'science_stu':[45,32,65,32,21]
})
```

```python
merged_df=pd.merge(df3,df4,on='row_no') # only merges common 
print(merged_df) 
```

### left, right,outer merge parameter how 

```python
left_merged=pd.merge(df3,df4,on='row_no',how='left') # left from row no df3 only prints
print(left_merged)
```

```python
right_merged=pd.merge(df3,df4,on='row_no',how='right') # right from row no df4 only prints
print(right_merged)
```

```python
outer_merged=pd.merge(df3,df4,on='row_no',how='outer') 
print(outer_merged)
```

# append -  concat 

```python
import pandas as pd
```

```python
df1=pd.DataFrame({
    'roll_no':[1,2,3,4,5],
    'maths':[98,29,76,54,65],
    'physics':[65,84,94,23,62]
})
df2=pd.DataFrame({
    'roll_no':[6,7,8,9,10],
    'maths':[48,99,56,64,85],
    'physics':[68,82,76,89,79]
})
```

```python
df3=pd.append([df1,df2]) #append is updated as concat
print(df3)
```

```python
df3=pd.concat([df1,df2])
print(df3)
```

```python
df4=pd.concat([df2,df1]) # 1st mentioned dataframe comes first
print(df4)
```

## ignore index

```python
df4=pd.concat([df2,df1],ignore_index=True) 
print(df4) # shows continuous index
```

```python
df4=pd.concat([df2,df1],ignore_index=True,sort=True) 
print(df4) # sorts based on alphabet from atoz in column name
```

```python
df1=pd.DataFrame({'roll_no':[1,2,3,4,5],'chemistry':[98,29,76,54,65],'physics':[65,84,94,23,62]})
df2=pd.DataFrame({'roll_no':[6,7,8,9,10],'maths':[48,99,56,64,85],'physics':[68,82,76,89,79]})
print(df1)
print(df2)
```

```python
df_combined=pd.concat([df1,df2],ignore_index=True)
print(df_combined)
```

# PIVOT TABLE PANDAS ### aggregate function

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\HP\Downloads\sample2.csv")
df.head()      
```

```python
# Creating Pivot Table to analyze average scores by Branch

pivot_branch = df.pivot_table(index='Branch',values=['Physics','Chemistry','Maths','Computer'],aggfunc='mean')
print('Average Scores by Branch:')
print(pivot_branch)
```

```python
# Creating Pivot Table to analyze average scores by Section

pivot_section = df.pivot_table(index='Section',values=['Physics','Chemistry','Maths','Computer'],aggfunc='mean')
print('Average Scores by Section:')
print(pivot_section)
```

```python
# creating Pivot Table to count the no of students per batch
pivot_count = df.pivot_table(index='Branch',values='Roll No.',aggfunc='count')
print('No of stu per Branch:')
print(pivot_count)
```

```python
# aggregate fn sum
pivot_section = df.pivot_table(index='Section',values=['Physics','Chemistry','Maths','Computer'],aggfunc='sum')
print('Average Scores by Section:')
print(pivot_section)
```

```python
# aggregate fn max
pivot_section = df.pivot_table(index='Section',values=['Physics','Chemistry','Maths','Computer'],aggfunc='max')
print('Average Scores by Section:')
print(pivot_section)
```
